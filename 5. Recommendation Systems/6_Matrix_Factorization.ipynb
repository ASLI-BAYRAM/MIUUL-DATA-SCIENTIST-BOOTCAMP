Boşlukları doldurmak için user'lar ve movie'ler için var olduğu varsayılan latent feature'ların(gizli özellikler) ağırlıkları var olan veri üzerinden bulunur ve bu ağırlıklar ile var olmayan gözlemler için tahmin yapılır.

* Bulunan ağırlıklar ile boş olan gözlemler doldurulur.

  ![Test](datasets/22.png)

Yukarıdaki görselde elimizde user ve movie'ler var bunların kesişimin de ise rating(r matrisi)yani puanlar var. Bu yöntem der ki kesişimindeki puanlar yaklaşık olarak kxn ve mxk matrisinden meydana gelir. Yani m tane kullanıcı n tane film k tane faktör(gizli faktör)'lerin çarpımında oluşan matrisler var.

User factors: Bir kullanıcı eğer bir filmi beğeniyorsa bu kullanıcının bu filmi beğenmesini ortaya çıkaran bazı sebepler vardır.Bu sebepler suan ölçülebilir değildir ama her bir kullanıcının vermiş oldugu puanı oluşturan gizli faktörlerdir.Örneğin biri film macera diye yüksek puan verebilir, başka biri oyuncusu için yüksek puan verebilir bunlar gizli faktörlerdir. 

Bu gizli faktörler filmler içerisinde de vardır.İşte ratingler yaklaşık olarak kullanıcı ve filmlerdeki gizli faktörlerin çarpışmasından ortaya çıkar.

* Rating matrisinin iki factor matrisin çarpımı(dot product) ile oluştuğu varsayılır.

* Factor matrisler: user latent factors, movie latent factors

* Latent factor: Gizli faktörler ya da değişkenler

* Kullanıcıların ve filmlerin latent feature'lar için skorlara sahip olduğu düşünülür.

* BU skorlar önce var olan veri üzerinden bulunur ve sonra BLANK bölümler bu ağırlıklara göre doldurulur.

* Latent faktörler şunlar olabilir; komedi, macera, aksiyon, gerilim filmleri gibi bir oyuncunun filmde olup olmaması, filmin uzunluğu vs.

  ![Test](datasets/23.png)

F1 ve F2 latent faktörler olsun. p ler kullanıcı açısından bu gizli faktörlerin ağırlıkları, q ise filmler acısından gizli faktörlerin ağırlıklarını temsil eder.Bu ağırlıkların çarpımından yaklaşık rating değeri bulunur.

  ![Test](datasets/24.png)

**Peki p ve q değerleri nasıl bulunacak?**

* Var olan değerler üzerinden iteratif şekilde tüm p ve q'lar bulunur sonra kullanılır.

* Başlangıçta rastgele p ve q değerleri ile rating matrisindeki değerler tahmin edilmeye çalışılır.

* Her iterasyonda hatalı tahminler düzenlenerek rating matristeki değerlere yaklaşılmaya çalışılır.

* Örneğin bir iterasyonda 5'e 3 dendiyse sonrakinde 4 sonrakinde 5 denir. ve hata minimuma indirilir.

* Böylece belirli bir iterasyon sonucunda p ve q matrisleri doldurulmuş olur.

| ------------------------------ | ----------------------------------------------------------------------------- |
| pki                      | Kullanıcı i'nin k. latent özelliği (örneğin: aksiyon filmi sevgisi)

| rij                      | Kullanıcı i'nin ürün j'ye verdiği gerçek puan (rating)

| gamma                    | Öğrenme oranı (learning rate), adım büyüklüğü                                 
| t                        | Güncelleme iterasyonundaki zaman (epoch sayısı gibi düşünülebilir)

<span style="color: blue; font-weight: bold;">3-MODEL TUNING</span>

```python
#maç: Modelin başarımını etkileyen parametreleri denemek.
param_grid = {'n_epochs': [5, 10, 20],#iteresyon sayısı
              'lr_all': [0.002, 0.005, 0.007]}
```

* n_epochs: Model kaç defa tüm veri seti üzerinden geçecek. Epoch sayısı arttıkça model daha fazla öğrenebilir ama overfitting riski de artar.

* lr_all: Öğrenme oranı (learning rate). Modelin parametrelerini güncellerken yaptığı adımın büyüklüğünü belirler.

```python
#Amaç: En iyi hiperparametre kombinasyonunu bulmak.
gs = GridSearchCV(SVD,
                  param_grid,
                  measures=['rmse', 'mae'],
                  cv=3,
                  n_jobs=-1,
                  joblib_verbose=True)
```

* SVD: Test edeceğimiz algoritma (Singular Value Decomposition).

* measures=['rmse', 'mae']: Hangi metriklere göre en iyisini arayacağız?

* rmse (Root Mean Squared Error): Hata karelerinin ortalamasının karekökü (tavsiye sistemlerinde en sık kullanılır).

* mae (Mean Absolute Error): Ortalama mutlak hata.

* cv=3: 3 katlı çapraz doğrulama (cross-validation).

* n_jobs=-1: Tüm işlemcileri kullan (hızlı sonuç alır).

* joblib_verbose=True: Hangi parametrelerin denendiği ve süreç çıktıları konsola yazılır.

```python
# Amaç: GridSearchCV tanımladığımız tüm parametre kombinasyonlarını deneyerek, en iyi sonucu bulsun.
gs.fit(data)
```

```python
#En iyi rmse skorunu ve o skoru getiren parametreleri öğren.
gs.best_score['rmse']
```

```python
gs.best_params['rmse']
```

<span style="color: blue; font-weight: bold;">3-FİNAL MODEL VE TAHMİN</span>

```python
#svd_model nesnesinin hangi özelliklere sahip olduğunu görmek. n_epochs, bu modelin kaç epoch boyunca eğitildiğini gösterir.
dir(svd_model)
svd_model.n_epochs#ilk modelin ön tanımlı değerlerine bakıyoruz
```

```python
# Amaç: GridSearchCV sonucu en iyi rmse’yi sağlayan parametrelerle yeni bir model oluşturmak.
svd_model = SVD(**gs.best_params['rmse'])#en iyi parametreleri modele ** kullanrak verdik
```

```python
# Amaç: Tüm veri üzerinde modeli yeniden eğitmek. Çünkü artık test ve eğitim ayrımı yapmıyoruz, elimizdeki tüm veriyle en iyi modeli oluşturmak istiyoruz.
data = data.build_full_trainset()#bütün veri setini train setine cevirdik.
```

```python
svd_model.fit(data)
```

```python
#Amaç: userId=1.0 olan kullanıcının movieId=541 olan filme vereceği puan tahminini üretmek.
svd_model.predict(uid=1.0, iid=541, verbose=True)
```
