```python
def check_df(dataframe, head = 5):
    print('##############SHAPE##############')
    print(dataframe.shape)
    print('##############TYPES##############')
    print(dataframe.dtypes)
    print('##############COLUMNS#####################')
    print(dataframe.columns)
    print('##############HEAD######################')
    print(dataframe.head(head))
    print('##############TAİL#####################')
    print(dataframe.tail(head))
    print('##############DESCRİBE##########################')
    print(dataframe.describe([0, 0.05,0.25, 0.50,0.75, 0.95, 0.99, 1]).T)
    print('##############NULL#####################')
    print(dataframe.isnull().sum())
check_df(df)
```

```python
def num_summary(dataframe, col_name, plot = False):
    quantiles = [0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 0.99 ]
    print(dataframe[col_name].describe(quantiles).T) #sayısal değişkenlerin betimsel istatistikleri
    print('################################')

    if plot:
        dataframe[col_name].hist()
        plt.xlabel(col_name)
        plt.title(col_name)
        plt.show()
```

<h3 style="text-align: center; color:red">DEĞİŞKENLERİN YAKALANMASI VE İŞLEMLERİN GENELLEŞTİRİLMESİ </h

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):#parametrelerin ön tanımlı değerlerini burda belirttik
    """
    Veri setindeki kategorik, numerik ve kategorik fakat kardinal degiskenlerin isimlerini verir.--
    Parameters
    ----------
    dataframe: dataframa
        degisken isimleri alınmak istenen dataframedir.
    cat_th: int, float
        numerik fakat kategorik olan degiskenler için sınıf eşik degeri
    car_th: int, float
        kategorik fakat kardinal degiskenler icin sınıf esik degeri

    Returns
    -------
    cat_cols: list
        kategorik degisken listesi
    num_cols: list
        numerik degisken listesi
    cat_but_car: list
        kategorik gorunumlu kardinal degisken listesi
    Notes
    -------
    cat_cols+num_cols+cat_but_car=toplam degisken sayısı
    num_but_cat cat_cols'un içerisindedir.
    Return olan 3 liste toplamı degisken sayısına eşittir: cat_cols +num_cols+cat_but_car
    """
    cat_cols = [col for col in dataframe.columns if str(dataframe[col].dtypes) in ['object', 'category', 'bool']]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].dtypes in ['int64', 'float64'] and dataframe[col].nunique() < cat_th]
    cat_but_car = [col for col in dataframe.columns if str(dataframe[col].dtypes) in ['object', 'category'] and dataframe[col].nunique() > car_th]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car] 

    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes  in ['int64', 'float64']] #bu sayısal değişkenleri verir
    num_cols = [col for col in num_cols if col not in cat_cols] 

    print(f'Observetions: {dataframe.shape[0]}')
    print(f'Variables: {dataframe.shape[1]}')
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')

    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(df)
```

```python
#Bonus
for col in df.columns:
    if df[col].dtypes == 'bool':
        df[col] = df[col].astype('int') #True false olarak gelen değişkenler 1/0 olarak görünür.adult male artık 1/0 görünüyor
```

```python
for col in cat_cols:
    cat_summary(df, col, True) 
```

<span style="color: blue; font-weight: bold;">Hedef Değişkenin Kategorik Değişkenler İle Analizi</span>

```python
def target_summary_with_cat(dataframe, target, cat_col):
    print(dataframe.groupby(cat_col).agg({target:'mean'}))
    print('####################################')
```

```python
for col in cat_cols:
    target_summary_with_cat(df, 'survived', col) #artık diğer kategorik değişkenleri survived açısından yorumlayabilriz
```

<span style="color: blue; font-weight: bold;">Hedef Değişkenin Sayısal Değişkenler İle Analizi</span>

```python
def target_summary_with_num(dataframe, target, num_col):
    print(dataframe.groupby(target).agg({num_col:'mean'}))
```

```python
for col in num_cols:
    target_summary_with_num(df, 'survived', col) 
```

<h3 style="text-align: center; color:red">KORELASYON ANALİZİ </h


Amacımız elimize bir veri seti geldiğinde ısı haritası aracılığıyla korelasyonlarına bakmak ve ihtiyacımız oldugunda yüksek korelasyonlu değişkenlerden bazılarını dışarda bırakabilmektir.

**Korelasyon**, değişkenlerin birbiri ile ilişksini ifade eden bir istatistiksel ölçümdür.-1 ile +1 arasında değerler alır.-1 ya da +1 ' e yaklaştıkça ilişkinin şiddeti kuvvetlenir. Eğer iki değişkenin arasındaki ilişki pozitif ise buna pozitif korelasyon denir ve bir değişkenin değeri arttıkça diğerininde değeri artar. İki değişken arasındaki ilişki negatif ise bir değişkenin değerleri artarken diğer değişkenin değeri azalır. Genellikle birbiri ile yüksek korelasyonlu olan değişkenlerin çalışmalarda bulunmamasını isteriz. Çünkü yüksek korelasyonlu bir ilişki var ise ikisi de aynı şeyi ifade ediyor demektir. Bu yüzden değişkenlerden birisini çalışma dışında bırakmak isteriz.

<span style="color: blue; font-weight: bold;">Yüksek Korelasyonlu Değişkenlerin Silinmesi</span>

```python
cor_matrix = df[num_cols].corr().abs()#tüm korelasyonunun mutlak değerini aldık
```

```python
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool_))
```

Bu satırın amacı, cor_matrix adındaki korelasyon matrisinin sadece üst üçgenini (upper triangle) elde etmektir. Yani matrisin alt kısmındaki tekrar eden değerleri maskeler (NaN yapar). Böylece simetrik tekrarları ortadan kaldırır, sadece bir kere değerlendirmeni sağlar.

np.ones(cor_matrix.shape)
Bu ifade, korelasyon matrisinle aynı boyutta (örneğin 3x3) sadece 1’lerden oluşan bir NumPy dizisi oluşturur.

np.triu(..., k=1)
np.triu() → upper triangle (üst üçgen) fonksiyonudur.
k=1 → ana köşegenin (diagonal) bir üstünden başlar. Yani köşegenin kendisini dahil etmez.

.astype(np.bool_)
Bu NumPy dizisini boolean (True/False) değerlerine çevirir.

cor_matrix.where(...)
.where(mask) → mask True olan yerlere dokunmaz, False olan yerleri NaN yapar.

Yani yukarıdaki boolean maskeye göre cor_matrix’in sadece üst üçgenindeki değerler korunur, diğerleri NaN olur.

     A     B     C
     
A  1.00  0.80  0.10

B  0.80  1.00  0.25

C  0.10  0.25  1.00

Uygulanan işlem sonucu:

     A     B     C     
A  NaN  0.80  0.10

B  NaN   NaN  0.25

C  NaN   NaN   NaN

```python
upper_triangle_matrix
```

```python
drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > 0.90)]
drop_list
```

Yukarıdaki kodun amacı korelasyon değeri 0.90'dan yüksek olan (çok benzer bilgi taşıyan) değişkenleri tespit etmek ve bunları silmek üzere bir listeye almak.

Bu sayede:

* Multicollinearity (çoklu doğrusal bağlantı) riski azaltılır.

* Model daha sade ve daha yorumlanabilir hale gelir.

* Aşırı benzer değişkenlerden doğabilecek gürültü (noise) engellenir.

upper_triangle_matrix[col] > 0.90 Bu ifade, o sütundaki değerlerin hangilerinin 0.90’dan büyük olduğunu True/False şeklinde döndürür.

any() fonksiyonu, bir dizide en az bir tane True varsa True döner.

```python
df.drop(drop_list, axis = 1, inplace =True) # veri setinden o değişkenleri sildik.
```

<span style="color: blue; font-weight: bold;">İşlemlerin Fonksiyonlaştırılması</span>

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr_matrix = dataframe.select_dtypes(include='number').corr().abs()
    upper_triangle_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]

    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.heatmap(corr_matrix, cmap="RdBu", annot=False)
        plt.title("Correlation Matrix")
        plt.show()

    return drop_list

drop_list = high_correlated_cols(df, plot=True)
drop_list
```

```python
high_correlated_cols(df.drop(drop_list, axis = 1), plot=True)#yüksek korelasyonlu değişkenleri çıkardık
```
