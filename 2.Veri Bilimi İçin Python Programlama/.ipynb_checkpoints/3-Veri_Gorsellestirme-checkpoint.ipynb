 <h1 style="text-align: center; color: RED">VERİ GÖRSELLEŞTİRME : MATPLOTLİB & SEABORN</h1> 
 

<span style="color: blue; font-weight: bold;">MATPLOTLIB</span>

Low level veri görselleştirme aracıdır. Yani seaborn'a göre daha çok çaba gerektirir.

***NOT***: Kategorik değişkenleri sütun grafikleri ile görselleştiriyoruz. Bunu da seaborn içerisindeki countplot ya da matplotlib içerisindeki bar plot ile gerçekleştirebiliriz.Sayısal değişkenleri ise histogram ve boxplot ile görselleştirebiliriz.

<span style="color: blue; font-weight: bold;">Kategorik Değişken Görselleştirme</span>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')
pd.set_option('display.width',500)#çıktının daha düzenli görünmesi için yapılan ayar
df.head()
```

```python
df['sex'].value_counts()# kategorik değişknelerle işlem yapmak istediğine aklına gelsin
```

```python
df['sex'].value_counts().plot(kind='bar')
plt.show()#grafiği ekrana çizdiridir.Sınıf dağılımların bilgisini sütun grafikle oluşturduk
```

df.plot(kind='bar', x='name', y='salary', title='Maaş Dağılımı', color='teal', figsize=(8,5), rot=45) bu fonksiyonun aldıgı argümanlar aşağıda belirtilmiştir.


* kind=	Grafik türü: 'line', 'bar', 'barh', 'hist', 'box', 'kde', 'area', 'pie', 'scatter' vs.
* x=	X eksenine karşılık gelen sütun adı (label)	x='name'
* y=	Y eksenine karşılık gelen sütun(lar) adı	y='salary'
* title=	Grafik başlığı	title='Maaş Dağılımı'
* xlabel / ylabel	= X/Y ekseni başlıkları	
* figsize= Grafik boyutu (genişlik, yükseklik)	
* color=	Renk veya renk listesi	
* legend=	Açıklama kutucuğu gösterilsin mi?	
* grid=	Grafik üzerine grid ekle	grid=True
* fontsize=	Etiket yazı boyutu	
* rot	= X ekseni etiketlerini döndür (derece cinsinden)	
* stacked	= Grupları üst üste mi göstersin?	

```python
df['sex'].value_counts().plot(xlabel = 'Cinsiyet Sınıfları', ylabel = 'Sınıf Sayıları', title = 'Cinsiyet', figsize = (3,3), color = 'red', grid = True ,kind='bar')
plt.show()#grafiği ekrana çizdiridir.Sınıf dağılımların bilgisini sütun grafikle oluşturduk
```

<span style="color: blue; font-weight: bold;">Sayısal Değişken Görselleştirme</span>

```python
plt.hist(df['age']) #sayısal değişkenlerin dağılım bilgisini verir.
plt.show()
```

plt.hist() fonksiyonunun aldığı argümanlar:


* x=	Verinin kendisi (genelde bir liste, NumPy array ya da Pandas Series)	
* bins=	Kaç aralık (kutucuk) olacağı ya da doğrudan aralıklar	
* range=	Histogramda gösterilecek minimum ve maksimum değer aralığı	range=(0, 100)
* density=	Frekans yerine olasılık yoğunluğunu gösterir (alan = 1 olur)	density=True
* weights =	Her verinin katkı ağırlığı	weights=np.ones_like(data)/len(data)
* cumulative=	Kümülatif (birikimli) histogram ister misin?	cumulative=True
* color=	Renk	color='skyblue'
* edgecolor=	Kutuların kenar çizgi rengi	edgecolor='black'
* alpha=	Saydamlık (0=şeffaf, 1=opak)	alpha=0.7
* label=	Efsane (legend) etiketi	label='Yaş Dağılımı'
* histtype=	Histogram tipi: 'bar', 'barstacked', 'step', 'stepfilled'


```python

plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black', alpha=1 )
plt.title('Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df['fare']) # kutu grafiktir.
plt.show()
```

```python
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Yaş Histogram + KDE')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show()
```

<span style="color: blue; font-weight: bold;">Matplotlib Özellikleri</span>



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot özelliği
x = np.array([1,8])
y = np.array([0,150])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'o')
plt.show()
```

```python
x = np.array([2,4,6,8,10])
y = np.array([1,3,5,7,9])

plt.plot(x, y, 'o') #noktalar kesişim noktalarını gösteriri
plt.show()
```

```python
#marker :işaretleyici özellikleri
y = np.array([13,28,11,100])
plt.plot(y, marker = '*', color ='r',linestyle = '-.')# markers = [o,*,., x,X, + , P, s, D, d, p,H,h]
plt.show()
```

```python
#multiple lines
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
#Labels
x  = np.array([80,85,90,95,100,105,110,115,120,125])
y  = np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x, y)
plt.title('Ana Başlık')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.grid()
plt.show()
```

```python
#SUBPLOTS

#1.plot
x  = np.array([80,85,90,95,100,105,110,115,120,125])
y  = np.array([240,250,260,270,280,290,300,310,320,330])

plt.subplot(1,2,1) #1 satırlık 2 sutunluk grafik oluşturdum
plt.title('1')
plt.plot(x, y)

#2.plot
plt.subplot(1,2,2) #1 satırlık 2 sutunluk grafik oluşturdum
plt.title('2')
plt.plot(x, y)
```

```python

#1.plot
x  = np.array([80,85,90,95,100,105,110,115,120,125])
y  = np.array([240,250,260,270,280,290,300,310,320,330])

#1. grafik
plt.subplot(1,3,1) #1 satırlık 3 sutunluk grafik oluşturdum
plt.title('1')
plt.plot(x, y)

#2. grafik
plt.subplot(1,3,2) 
plt.title('2')
plt.plot(x, y)

#3. grafik
plt.subplot(1,3,3) 
plt.title('3')
plt.plot(x, y)
```

<span style="color: blue; font-weight: bold;">SEABORN İLE VERİ GÖRSELLEŞTİRME</span>

Bir veri görselleştirme kütüphanesidir. Yüksek seviye bir kütüphanedir yani az çaba ile çok iş yapma imkanı sağlar.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x = df['sex'], data = df) #kategorik değişken görselleştirme
plt.show()
```

```python
sns.boxplot(x = df['total_bill'], data=df, color = 'r')
```

```python
df['total_bill'].hist()# bu bir pandas fonksiyonudur.
plt.show()
```
