<span style="color: blue; font-weight: bold;">GÖREV16:</span> 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 vericek bir fonksiyon yazınız.Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz oluşturunuz. (apply ve lambda yapılarını kullanınız)

```python
df['age_flag'] = df['age'].apply(lambda x: 1.30 if x < 30 else 0 )
df.head()
```

```python
#Fonksiyonlu yöntem
def age_30(age):
    if age < 30:
        return 1.30
    else:
        return 0
df['age_flag']  = df['age'].apply(lambda x :  age_30(x) )
```

<span style="color: blue; font-weight: bold;">GÖREV17:</span> Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.

<span style="color: blue; font-weight: bold;">GÖREV18:</span> Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill  değerlerinin toplamını, min, max ve ortalamasını bulunuz.

<span style="color: blue; font-weight: bold;">GÖREV19:</span> Günlere ve time göre total_bill değerlerinin toplamını, min, max ve ortalamasını bulunuz.

```python
df.groupby(['day','time']).agg({'total_bill': ['min','max','mean']})
```

<span style="color: blue; font-weight: bold;">GÖREV20:</span> Lunch zamanına ve kadın müşterilere ait total_bill ve tip  değerlerinin day'e göre toplamını, min, max ve ortalamasını bulunuz.

```python
df[(df["time"] == "Lunch") & (df["sex"] == "Female")].groupby("day").agg({"total_bill": ["sum","min","max","mean"],
                                                                           "tip":  ["sum","min","max","mean"]})
```

<span style="color: blue; font-weight: bold;">GÖREV21:</span> size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir?

```python
df.loc[(df['size'] < 3) & (df['total_bill'] > 10), 'total_bill'].mean()
```

<span style="color: blue; font-weight: bold;">GÖREV22:</span> total_bill_tip_sum adında yeni bir değişken oluşturun. Her bir müşterinin ödediği totalbill ve tip in toplamını versin.
