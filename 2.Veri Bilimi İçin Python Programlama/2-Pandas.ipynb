<span style="color: blue; font-weight: bold;">Toplulaştırma & Gruplama(Aggregation & Grouping)</span>

Toplulaştırma bir veri yapısının içinde barınan değerleri toplu bir şekilde temsil etmek demektir. count(), first(), last(), mean(), meadian(), min(), max(), std(), sum() bunlar toplulaştırma fonksiyonlarıdır. Bunlar tek başına da kullanılır. Fakat genellikle toplulaştırma ve gruplama işlemleri hep bir arada olur.



```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')

#Amacımız kadın ve erkeklere göre yaş ortalaması almak
print(df['age'].mean()) #bu bize tüm yaşların ortalamasını verir. Biz cinsiyete göre ortalama almak istiyoruz
```

```python
print(df.groupby('sex')['age'].mean())#cinsiyete göre grupladık ve yaş değişkeninn ortalamasını aldık
```

```python
df.groupby('sex').agg({'age':'mean'})#yukarıdaki işlemin farklı verisyonu
```

```python
df.groupby('sex').agg({'age':['mean', 'sum']}) #birden fazla fonk. kullandık
```

```python
df.groupby('sex').agg({'age':['mean', 'sum'],
                       'embarked':'count'})#burda embarked'ın değil cinsşyet sayısını verdi
```

```python
df.groupby(['sex','embark_town']).agg({'age':['mean', 'sum'],
                                     'survived':'mean'})# birden fazla kategorik değişkene göre kırılım yapmak
```

```python
df.groupby(['sex','embark_town', 'class']).agg({'age':['mean', 'sum'],
                                                'survived':'mean',
                                                'sex':'count'})
```

<span style="color: blue; font-weight: bold;">Apply & Lambda</span>

Apply satır ya da sutünlarda otomatik olarak fonksiyon çalıştırma imkanı sağlar. 
Lambda bir fonksiyon tanımlama şeklidir. Kullan at fonksiyondur, yani kod akışı esnasında bir kere kullanıp atayım daha sonra bu fonksiyona ihtiyacım yok demektir.

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('titanic')
```

```python
df['age2']  = df['age']**2
df['age3'] = df['age']*5

```

```python
#amacımız bu yaş değişkenlerin hepsini 10' bölmek. Bunu tek tek yapacağımıza göre klasik yolla döngü yazalım
for col in df.columns:
    if 'age' in col:
        df[col] = df[col] / 10
print(df.head())
```

```python
#apply ve lambda ile yapma
df[['age','age2', 'age3']].apply(lambda x: x/10).head()

#değişkenleri tek tek yazmamak için
df.loc[:, df.columns.str.contains('age')].apply(lambda x: x/10).head()
```

df.columns.str.contains('age') : 

* df.columns.str
Bu, Pandas’ın Index objesi üzerinde string işlemlerini uygulamak için kullanılan bir özellik. .str sayesinde sütun adlarına Python string fonksiyonlarını topluca uygulayabilirsin.

* df.columns.str.contains('age')
Bu ifade, sütun isimleri içinde 'age' kelimesini içerenleri True, içermeyenleri False yapar.

```python
df.loc[:, df.columns.str.contains('age')].apply(lambda x: (x -x.mean()) / x.std()).head()
```

```python
#Apply fonksiyonunu başka fonksiyonlarla da kullanabliriz

def scaler(col_name):
    return (col_name - col_name.mean()) / col_name.std()
#kalıcı olması için atama yaptım
df.loc[:, df.columns.str.contains('age')] = df.loc[:, df.columns.str.contains('age')].apply(scaler).head()
df.head()
```

<span style="color: blue; font-weight: bold;">Birleştirme İşlemleri</span>

Concat: ön tanımlı değeri axis=0'dır yani satırlara göre alt alta birleştirir. Eğer yana yana birleştirmek istersek axis = 1 dememiz yeterli olacaktır.

Merge: Ortak değerler üzerinden birleşitirir.

```python
import numpy as np
import pandas as pd

#rastgele 2 dataframe oluşturduk.
m = np.random.randint(1, 30, size=(5,3))
df1 = pd.DataFrame(m, columns = ['var1','var2','var3'])
df2 = df1 + 99 
```

```python
print(df1)
print(df2)
```

```python
#İki dataframe alt alta birleştirmek istersek:
pd.concat([df1, df2])
```

```python
#İki dataframe alt alta birleştirmek istersek:
pd.concat([df1, df2], ignore_index = True) #indexleri baştan sıraladı
```

```python
df1 = pd.DataFrame({'employees' : ['john','dennis','mark','maria'],
                   'group':['accounting','engineering','engineering','hr']})
df2 = pd.DataFrame({'employees' : ['mark','john','dennis','maria'],
                   'start_date':[2010, 2009, 2014, 2019]}) 
```

```python
df3 = pd.merge(df1, df2, on = 'employees') #ortak değişken üzerinden birleştirdik
df3
```

```python
#Her çalışanın müdür bilgisine erişmek
df4 = pd.DataFrame({'group':['accounting','engineering','hr'],
                   'manager':['Caner','Mustafa','Berkcan']}) 
```

```python
pd.merge(df3, df4, on = 'group')
```
