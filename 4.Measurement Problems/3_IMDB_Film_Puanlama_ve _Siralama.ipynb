<span style="color: blue; font-weight: bold;">2-IMDB AĞIRLIKLI DERECELENDİRME</span>

IMDB'nin 2015 ' kadar kullandığı formul:

weighted_rating = (v/(v + M)*r) + (M/(v+M)*C)

r = vote average

v = vote count

M = minimum votes required to be listed in the top 250(gerekli oy sayısı)

C = the mean vote across the whole report(currently 7.0)(genel kitle ortalaması)

```python
M = 2500
C = df['vote_average'].mean()
```

```python
def weighted_rating(r, v, M, C):
    return (v/(v + M)*r) + (M/(v+M)*C)
```

```python
df['weighted_rating'] = weighted_rating(df['vote_average'], df['vote_count'], M, C)
```

```python
df.sort_values('weighted_rating', ascending=False).head(20)
```

<span style="color: blue; font-weight: bold;">3-BAYESIAN AVERAGE İLE SIRALAMA </span>

```python
df_ = pd.read_csv("C:\\Users\\ASLI\\Desktop\\Data Scientist Bootcamp\\4.Measurement Problems\\datasets\\imdb_ratings.csv",index_col=0)
```

```python
df_.head()#filmlerin aldıgı yıldız sayıları bu veri setinde mevcut
```

```python
def bayesian_average_rating(n, confidence=0.95):
    if sum(n) == 0:
        return 0

    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)

    first_part = 0.0
    second_part = 0.0

    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n_k + 1) / (N + K)
        second_part += (k + 1)**2 * (n_k + 1) / (N + K)

    score = first_part - z * math.sqrt((second_part - first_part**2) / (N + K + 1))
    return score
```

```python
df_['bar_score'] = df_.apply(lambda x: bayesian_average_rating(x[['one','two','three','four','five',
                                                               'six','seven','eight','nine','ten']]), axis=1)
```

```python
df_.sort_values('bar_score',ascending=False).head(20)
```
