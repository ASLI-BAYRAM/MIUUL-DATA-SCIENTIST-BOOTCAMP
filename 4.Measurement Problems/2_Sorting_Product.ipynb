<span style="color: blue; font-weight: bold;">4-BAYES ORTALAMA DERECELENDİRME PUANI(BAYESİAN AVERAGE RATİNG SCORE)</span>

Bu yöntem ile puanların dağılım bilgisini kullanarak olasılıksal bir ortalama hesaplayacaz.

```python
def bayesian_average_rating(n, confidence=0.95):
    if sum(n) == 0:
        return 0

    K = len(n)
    z = st.norm.ppf(1 - (1 - confidence) / 2)
    N = sum(n)

    first_part = 0.0
    second_part = 0.0

    for k, n_k in enumerate(n):
        first_part += (k + 1) * (n_k + 1) / (N + K)
        second_part += (k + 1)**2 * (n_k + 1) / (N + K)

    score = first_part - z * math.sqrt((second_part - first_part**2) / (N + K + 1))
    return score
```

**Fonksiyonun Amacı:**

Kullanıcı puanlarının adaletsiz dağılımına karşı (örneğin sadece 1 kişi 5 yıldız verdiyse vs.) daha adil bir "ortalama puan" üretmek.

**def bayesian_average_rating(n, confidence=0.95):**

n: Kullanıcı puanlarının histogramı gibi düşün. Örneğin 5 yıldızlı bir sistemde:

n = [20, 15, 40, 100, 150]

burada sırasıyla 1 yıldız alan 20 oy, 2 yıldız alan 15 oy... 5 yıldız alan 150 oy demektir.

confidence: Güven düzeyi (varsayılan %95). İstatistiksel güven aralığını hesaplarken kullanılır.

**if sum(n)==0:
        return 0**

Eğer hiç oy verilmemişse, ortalama hesaplamak anlamlı değildir. Bu nedenle sıfır döner.

**K = len(n)**

K: Kaç farklı puan seviyesi var (örn. 5 yıldız sistemi -> K = 5)


**z = st.norm.ppf(1-(1-confidence) / 2**

z: Standart normal dağılıma göre z-skoru.

%95 güven düzeyi için z ≈ 1.96

st.norm.ppf(...) ile scipy.stats kütüphanesinden elde edilir.

**N = sum(n)**

N: Toplam oy sayısı

**Ortalama ve Varyans Hesabı**
first_part=0.0

second_part = 0.0

Bu iki değişken, BAR formülünün beklenen değeri (mean) ve varyansı için kullanılır.

**for k,n_k enumerate(n):
        first_part += (k+1)*(n[k]+1) /(N+K)
        second_part += (k+1)*(k+1)*(n[k]+1) /(N+K)**
        
numerate(n) sayesinde hem indeks k hem değer n_k alınır.

(k+1) → çünkü Python'da index 0'dan başlar ama yıldızlar 1'den (1-5) başlar.

n[k]+1: Laplace smoothing (her sınıfa 1 oy ekleyerek sıfırları engellemek için)

(N + K): Toplam oy sayısına K tane 1 eklenmiş olur. Bu da Laplace smoothing'in toplamı.

**score = first_part - z * math.sqrt((second_part - first_part^^2)/(N + K + 1))**

first_part: Ortalama puan (smoothed)

second_part: Varyansla ilgili terim

z * √(...): Güven aralığı

score: Ortalama puan - güvenlik payı → bu sayede çok az oy almış ama çok yüksek puanlı ürünlerin sıralamada haksızca yukarı çıkması engellenir.

```python
df['bar_score'] = df.apply(lambda x: bayesian_average_rating(x[['1_point',
                                                                '2_point',
                                                                '3_point',
                                                                '4_point',
                                                                '5_point']]),axis=1)
```

<span style="color: blue; font-weight: bold;">5-KARMA SIRALAMA</span>

####################

Hybrid Sorting: BAR Score + Diğer Faktorler

####################

1-Rating Products
 - Average
 - Time-Based Weighted Average
 - User-Based Weighted Average
 - Weighted Rating
 - Bayesian Average Rating Score

2-Sorting Products
 - Sorting by Rating
 - Sorting by Comment Count or Purchase Count
 - Sorting by Rating, Comment and Purchase
 - Sorting by Bayesian Average Rating Score (Sorting Products with 5 Star Rated)
 - Hybrid Sorting: BAR Score + Diğer Faktorler

```python
def hybrid_sorting_score(dataframe, bar_w = 0.6, wss_w=0.4):
    bar_score = dataframe.apply(lambda x: bayesian_average_rating(x[['1_point',
                                                                '2_point',
                                                                '3_point',
                                                                '4_point',
                                                                '5_point']]),axis=1)
    wss_score = weighted_sorting_score(dataframe)

    return bar_score * bar_w + wss_score*wss_w
```

```python
df['hybrid_sorting_score']=hybrid_sorting_score(df)
```

```python
df.sort_values('hybrid_sorting_score', ascending=False).head(20)
```

Hybrid yaklaşımda satın alma, yorum sayısı, aldığı puanlar dikkate alınarak bir sıralama yapıldı.
